```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

np.random.seed(42)


num_samples = 500


room_count = np.random.randint(1, 6, size=num_samples)


square_meter = np.random.randint(50, 201, size=num_samples)


house_age = np.random.randint(0, 31, size=num_samples)


locations = ['Mahalle_A', 'Mahalle_B', 'Mahalle_C', 'Mahalle_D']
location = np.random.choice(locations, size=num_samples)


base_price = 50000
price = base_price + (square_meter * 150) - (house_age * 1000) + (room_count * 5000) + np.random.randint(-20000, 20000, size=num_samples)


data = pd.DataFrame({
    'Oda Sayısı': room_count,
    'Metrekare': square_meter,
    'Ev Yaşı': house_age,
    'Lokasyon': location,
    'Fiyat': price
})


le = LabelEncoder()
data['Lokasyon'] = le.fit_transform(data['Lokasyon'])


data.to_csv('veri.csv', index=False)


data.head()
```
